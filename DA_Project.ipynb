```python
import pandas as pd
from google.colab import files
import gdown

file_id = '1gSq7VXdW_goawIg-yxtmJiBGL_WNyTf0'

url = f'https://drive.google.com/uc?id={file_id}'

output = 'ai_job_dataset.csv'
gdown.download(url, output, quiet=False)

df = pd.read_csv(output)
df.head()
```

```python
import pandas as pd
import numpy as np
#Variable Objetivo
display('------------Variable Objetivo-------------')

variable_objetivo = 'salary_usd'
display(df[variable_objetivo].head())

#Estadisticas descriptivas

display('------------Estadisticas Descriptivas-------------')
estadisticas_basicas = df.describe()
display(estadisticas_basicas)


#Convertir variables categoricas a dicotomaticas
#Niveles de experiencia



display('------------Conversion / Numerico-------------')
diccionario_niveles_Experiencia = {
    'EN': 1,
    'MI': 2,
    'SE': 3,
    'EX': 4
}

diccionario_tipo_empleo = {
    'FT': 1,
    'PT': 2,
    'CT': 3,
    'FL': 4
}

diccionario_tamaño_empresa = {
    'S': 1,
    'M': 2,
    'L': 3
}

diccionario_educacion = {
    'Bachelor': 1,
    'Associate': 2,
    'Master': 3,
    'PhD': 4
}




df['experience_level'] = df['experience_level'].str.strip().str.upper()
df['experience_level'] = df['experience_level'].map(diccionario_niveles_Experiencia)
display(df[['job_title','experience_level']].head())


df['employment_type'] = df['employment_type'].str.strip().str.upper()
df['employment_type'] = df['employment_type'].map(diccionario_tipo_empleo)
display(df[['job_title','employment_type']].head())

df['company_size'] = df['company_size'].str.strip().str.upper()
df['company_size'] = df['company_size'].map(diccionario_tamaño_empresa)
display(df[['job_title','company_size']].head())


df['education_required'] = df['education_required'].str.strip()
df['education_required'] = df['education_required'].map(diccionario_educacion)
display(df[['job_title','education_required']].head())

df['remote_ratio'] = df['remote_ratio']
display(df[['job_title','remote_ratio']].head())

display('------------Verificar Datos Null o NaN-------------')
df.isnull().sum()
```

```python
import pandas as pd
import plotly.express as px

# Agrupar por localización y calcular salario promedio
grouped = df.groupby('company_location')['salary_usd'].agg(['mean', 'count']).reset_index()
grouped.rename(columns={'mean': 'avg_salary_usd', 'count': 'count_jobs'}, inplace=True)

# Crear gráfico de burbujas
fig = px.scatter(
    grouped,
    x='company_location',
    y='avg_salary_usd',
    size='count_jobs',
    color='company_location',
    size_max=60,
    title='Gráfica de Burbujas: Promedio de salario por localización',
    labels={
        'avg_salary_usd': 'Salario Promedio (USD)',
        'count_jobs': 'Cantidad de trabajos'
    }
)

fig.show()
```

```python
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
%matplotlib inline
```

```python
lm = LinearRegression()
lm
```

```python
X = df[['experience_level']]
y = df['salary_usd']
lm.fit(X,y)
lm_coef_deter = lm.score(X,y)
lm_intercepto = lm.intercept_
lm_coeficiente = lm.coef_
print(f"Coeficiente de Determinación = ",{lm_coef_deter})
print(f"Intercepto = ",{lm_intercepto})
print(f"Coeficiente = ", str(lm_coeficiente))
```

```python
Z = df[['experience_level', 'company_size', 'remote_ratio', 'benefits_score']]
lm.fit(Z, df['salary_usd'])
lm1_coef_deter = lm.score(Z,y)
lm1_intercepto = lm.intercept_
lm1_coeficiente = lm.coef_
print(f"Coeficiente de Determinación = ",{lm1_coef_deter})
```

```python
import plotly.express as px

#Representación visual de la regresión lineal multiplel
y_pred = lm.predict(Z)

plt.figure(figsize=(8,6))
sns.scatterplot(x=df['salary_usd'], y=y_pred)
plt.xlabel("Salario Real (USD)")
plt.ylabel("Salario Predicho (USD)")
plt.title("Salario Real vs Predicho - Modelo de Regresión Múltiple")
plt.plot([df['salary_usd'].min(), df['salary_usd'].max()],
         [df['salary_usd'].min(), df['salary_usd'].max()],
         color='red', linestyle='--')  # línea 45° (predicción perfecta)
plt.show()
```

```python
#MODELO MLR
yhat = lm.predict(Z)
residuos = df['salary_usd'] - yhat

plt.figure(figsize=(10,6))
sns.residplot(x=df['experience_level'], y=residuos, lowess=True, line_kws={'color': 'red'})
plt.xlabel("Nivel Experiencia") # Label the x-axis correctly based on what you plot
plt.ylabel("Residuos")
plt.title("Residuos vs Nivel experiencia (from MLR model)") # Update title for clarity
plt.show()
```
